## Project map
This project practices the first layer of the course: image tensors, classification, localization metrics, segmentation overlap, and temporal identity.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

np.random.seed(7)
digits = load_digits()
X_images = digits.images
y = digits.target
X = X_images.reshape(len(X_images), -1)

print(X_images.shape, X.shape, y.shape)
```

```python
fig, axes = plt.subplots(4, 5, figsize=(9, 6))
for ax, img, label in zip(axes.ravel(), X_images[:20], y[:20]):
    ax.imshow(img, cmap="gray")
    ax.set_title(f"label={label}")
    ax.axis("off")
plt.tight_layout()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=7, stratify=y
)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

clf = LogisticRegression(max_iter=800, solver="lbfgs", multi_class="auto")
clf.fit(X_train_s, y_train)
y_pred = clf.predict(X_test_s)
print("Accuracy:", np.mean(y_pred == y_test))
```

```python
def accuracy(y_true, y_pred):
    return np.mean(np.asarray(y_true) == np.asarray(y_pred))

print("Manual accuracy:", accuracy(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot(cmap="Blues", values_format="d")
plt.title("Confusion matrix")
plt.show()
```

```python
mask_a = np.zeros((32, 32), dtype=bool)
mask_b = np.zeros((32, 32), dtype=bool)
mask_a[8:24, 9:24] = True
mask_b[11:27, 13:28] = True

def iou(pred, gt):
    intersection = np.logical_and(pred, gt).sum()
    union = np.logical_or(pred, gt).sum()
    return intersection / union

def dice(pred, gt):
    intersection = np.logical_and(pred, gt).sum()
    return 2 * intersection / (pred.sum() + gt.sum())

print("IoU:", iou(mask_b, mask_a))
print("Dice:", dice(mask_b, mask_a))

fig, axes = plt.subplots(1, 3, figsize=(10, 3))
axes[0].imshow(mask_a, cmap="gray"); axes[0].set_title("Ground truth")
axes[1].imshow(mask_b, cmap="gray"); axes[1].set_title("Prediction")
axes[2].imshow(mask_a.astype(int) + mask_b.astype(int), cmap="viridis"); axes[2].set_title("Overlap")
for ax in axes: ax.axis("off")
plt.show()
```

```python
centers = []
for t in range(10):
    x = 5 + 2.5 * t
    y_pos = 8 + 1.5 * t
    centers.append((x, y_pos))
centers = np.array(centers)

plt.figure(figsize=(5, 5))
plt.plot(centers[:, 0], centers[:, 1], marker="o")
for t, (x, y_pos) in enumerate(centers):
    plt.text(x + 0.2, y_pos, str(t))
plt.gca().invert_yaxis()
plt.title("Simple tracking: centroid over time")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
